### Предварительная подготовка:
В фале находился скрипт для создания БД. Запустил скрипт и выгрузил в csv файл outlets_h.csv

````python
import pandas as pd
import re
import numpy as np
````

````python
data = pd.read_csv('outlets_h.csv', sep=';', encoding ='mbcs')
data.head(50)
````

````python
data['name_nwe'] = data['Торг_точка_грязная']
data['adr_nwe'] = data['Торг_точка_грязная_адрес']
````

````python
data_re = data
````

````python
# Избавимся в столбцах с адресом и названия точек от спец символов уберем лишние пробелы и приведем все к нижнему регистру:
data_re['name_nwe'] = data_re['name_nwe'].apply(lambda x: re.sub('\W+',' ', x ) )
data_re['adr_nwe'] = data_re['adr_nwe'].apply(lambda x: re.sub('\W+',' ', x ) )
data_re['name_nwe'] = data_re['name_nwe'].str.lower()
data_re['adr_nwe'] = data_re['adr_nwe'].str.lower()
data['name_nwe'] = data['name_nwe'].str.strip()
data['adr_nwe'] = data['adr_nwe'].str.strip()
````

````python
data_re.head(45)
````

````python
data_re[['name_nwe','adr_nwe']].duplicated().sum()
````

Уже имеем 5147 явных дубликатов.

Было бы хорошо уточнить у составителя таблицы как она заполнялась, в какой последовательности. Возможно дало бы нам более ясное понимание что означает "-" в строках с адресом и по какому принципу заполнялось поле адреса с "он же".

За неимением такой информации предположим что "-" это если адрес совсем отсутствует, а "он же" актуален для точек расположенных выше в файле.

````python
for i in range(len(data_re)):
    if data_re['adr_nwe'][i] != 'он же':
        z = data_re['adr_nwe'][i]
    else:
       data_re['adr_nwe'][i] = z 
````

Можем наблюдать что дубликатов названий гораздо больше чем названий в связке с адресами, при получении информации о заполнении таблицы можно будет увеличить эффективность практически в 2 раза. Без нее мы можем не корректно заполнить адреса для некоторых точек. В данном случае считаю что отсутсвсие информации лучше чем не корректная.

````python
data_re['dub_y'] = data_re[['name_nwe','adr_nwe']].duplicated() 
````

````python
# в датафреме ниже мы получили явные дубликаты, их гораздо меньше чем на ссамом деле но это практически без потери либо
# с отстутсвием не корректной информации
data_re.head(45)
````

Если потеря и искажение данных не так критично то можем проделать следующее:

Предполагая что если в столбце адреса стоит "он же" - это дубликат предыдушей записи не смотря на название органаизации.
Так же дубликатами будем считать одинаковые названия ИП и одинаковые адреса.



````python
data_re['dub_name'] = data['name_nwe'].duplicated()
data_re['dub_adr'] = data['adr_nwe'].duplicated()
data_re['dub_same'] = data['Торг_точка_грязная_адрес'] == 'он же'
````

````python
data_re.head(10)
````

Далее будем считать если хотя бы в одном столбце есть метка что строка является дубликатом будем считать что это дубликат

Либо если изменятся входные данные можем использовать разные сочетания данных столбцов для определения дубликатов и изменить результат.

````python
data_re['dub_total'] = data_re['dub_y']+data_re['dub_name']+data_re['dub_adr']+data_re['dub_same']
````

````python
data_re.head()
````

````python
len(data_re) - data_re['dub_total'].sum()
````

Получаем всего 2848 уникальных значений.

````python
data_re['outlet_clean_id'] = range(len(data_re))
idx = data_re.query("dub_total == True").index
data_re.loc[idx, 'outlet_clean_id'] = np.NaN
````

````python
data_re.columns
````

````python

outlets = data_re[['id', 'Город дистрибьютора', 'Торг_точка_грязная',
       'Торг_точка_грязная_адрес', 'outlet_clean_id', 'name_nwe', 'adr_nwe']]
outlets.to_csv('outlets', encoding='utf-8' )
outlets.head(10)
````

````python
outlets_clean = data_re[~data_re['outlet_clean_id'].isna()][['adr_nwe','outlet_clean_id']]
outlets_clean.rename(columns={'adr_nwe':'Торг_точка_чистый_адрес','outlet_clean_id':'id'})
outlets_clean.to_csv('outlets_clean', encoding='utf-8' )
````
